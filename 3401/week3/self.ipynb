```python
import cv2
import numpy as np
```

```python
img = cv2.imread('../image/lena.jpg')
print(f"img type : {type(img)}")
print(f"img shape : {img.shape}")
```

```python
def showImage(img) :
    cv2.imshow("xxxx",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    for i in range (1,5):
        cv2.waitKey(1)
```

```python
showImage(img)
```

```python
img = img.astype(np.float16)
```

```python
gray_img = (0.1140 * img[:,:,0]) + (0.5870 * img[:,:,1]) + (0.2989 * img[:,:,2])
gray_img = gray_img.astype(np.uint8)
showImage(gray_img)
```

```python
negative_img = 255 - gray_img
showImage(negative_img)
```

```python
double_negative_img = 255 - negative_img
showImage(double_negative_img)
```
