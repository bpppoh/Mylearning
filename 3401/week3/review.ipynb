```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def showImage(image):
    cv2.imshow("Image", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    for i in range(5):
        cv2.waitKey(1)
```

cv2.imread("filepath")

will return numpy array

```python
img = cv2.imread("../image/lena.jpg")
gamma = 0.5
img_float = img.astype(np.float16)
img_new = (img_float/img_float.max()) ** gamma * 255
img_new = img_new.astype(np.uint8)
showImage(img)
showImage(img_new)
```
