```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../image/lena.jpg',1)
print(image)
```

```python
def showImage(image) :
    cv2.imshow('header',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    for i in range(5):
        cv2.waitKey(2)
```

```python
showImage(image)
showImage(image[:,:,0])
showImage(image[:,:,1])
showImage(image[:,:,2])
```

```python
plt.imshow(image[:,:,::-1])
plt.show()
plt.imshow(image[:,:,0] , cmap='Blues')
plt.show()
plt.imshow(image[:,:,0] , cmap='Greens')
plt.show()
plt.imshow(image[:,:,0] , cmap='Reds')
plt.show()
```
