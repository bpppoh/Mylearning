```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def imageShow(image) :
    cv2.imshow('xxxx',image)
    cv2.waitKey()
    cv2.destroyAllWindows()
    for i in range(0,5) :
        cv2.waitKey(1)
```

```python
img = cv2.imread('../image/dark2.png',0)

plt.figure(figsize=(15,15))

plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')
plt.title('original image')
imageShow(img)

plt.subplot(2,2,2)
plt.hist(img.ravel() , 256 , [0,256] , color='black')
plt.title('Histogram (Pixel Intensity Distribution)')
plt.xlabel('Intensity')
plt.ylabel('pixels amount')

gamma = 1.0
img_float = img.astype(np.float16)
img_new = (img_float / img.max()) ** gamma * 255
img_new = img_new.astype(np.uint8)

plt.subplot(2,2,3)
plt.title('New image')
plt.imshow(img_new , cmap='gray')
imageShow(img_new)

plt.subplot(2,2,4)
plt.hist(img_new.ravel() , 256 , [0,256] , color='black')
plt.title('Histogram (Pixel Intensity Distribution)')
plt.xlabel('Intensity')
plt.ylabel('pixels amount')

plt.show()
```
