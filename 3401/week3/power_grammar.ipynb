```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def showImage(image) :
    cv2.imshow('xxxx',image)
    cv2.waitKey()
    cv2.destroyAllWindows()
    for i in range(0,5) :
        cv2.waitKey(1)
```

```python
img = cv2.imread('../image/dark2.png',0)
showImage(img)
```

```python
gamma = 0.5
img = img.astype(np.float16)
img_new = (img / img.max()) ** gamma * 255
img_new = img_new.astype(np.uint8)
showImage(img_new)
```
