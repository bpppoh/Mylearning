```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def showImage(img):
    cv2.imshow("image",img) 
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    for i in range(3):
        cv2.waitKey(1)
```

```python
img = cv2.imread("../image/dark2.png")
print(type(img))
showImage(img)
```

```python
gamma = 0.3

#สร้าง array ของค่าสีเมื่อถูกคำนวณด้วย gamma แล้ว
gammaTable = np.array([i/255.0 for i in range(256)])
gammaTable = np.power(gammaTable, gamma)

#แปลงค่าสีให้เป็น uint8
gammaTable = np.uint8(gammaTable * 255) 

#ใช้ LUT เพื่อเทียบค่าสีระหว่างภาพดั้งเดิมกับค่าสีที่ถูกคำนวณด้วย gamma
img_new = cv2.LUT(img, gammaTable)

img_concat = cv2.hconcat([img, img_new])
showImage(img_concat)
```
