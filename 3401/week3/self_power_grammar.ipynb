```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def showImage(image) :
    cv2.imshow("Image",image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    for i in range(5):
        cv2.waitKey(1)
```

```python
img = cv2.imread("../image/dark2.png")
showImage(img)
gamma = 0.3
img_new = img.astype(np.float64)
img_new = (img_new/img_new.max()) ** gamma * 255
img_new = img_new.astype(np.uint8)
cv2.imwrite("../image/dark2_gamma.png", img_new)
```
