```python
import cv2
import numpy as np
```

```python
img = cv2.imread("../image/RGB_test.png")
img_gray = cv2.imread("../image/RGB_test.png" , cv2.IMREAD_GRAYSCALE)
print(type(img))
```

```python
def showImage(img) :
    cv2.imshow("xxxx",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    for i in range (1,5):
        cv2.waitKey(1)
```

```python
print(type(img),img.shape)
b,g,r = cv2.split(img)
print(type(b) , b.shape)
print(type(g) , g.shape)
print(type(r) , r.shape)
```

```python
img2 = cv2.vconcat([b,g,r])
showImage(img2)
```

```python
print(img_gray.shape , type(img_gray.shape))
showImage(img_gray)
row , col , n = img.shape
```

```python
gray = np.zeros([row,col] , dtype = 'float')
print(gray)
```

```python
gray = gray.astype(np.float16)
for i in range(row) :
    for j in range(col) :
        gray[i,j] = 0.1140*img[i,j,0] + 0.5870*img[i,j,1] + 0.2989*img[i,j,2]
```

```python
gray = gray.astype(np.uint8)
```

```python
showImage(gray)
```
