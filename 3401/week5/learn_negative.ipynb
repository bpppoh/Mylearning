```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def imgShow(img) :
    cv2.imshow("image",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    for i in range(1) :
        cv2.waitKey(1)
```

```python
img = cv2.imread("../image/i1552-5775-14-3-74-f01.jpg",0)
print(type(img))
negImg = 255 - img
```

```python
plt.figure(figsize=(15,10))

plt.subplot(1,2,1)
plt.imshow(img , cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(negImg , cmap='gray')
plt.axis('off')

plt.show()
```

```python
colorImg = cv2.imread('../image/lena.jpg')[:,:,::-1]
negColorImg = 255 - colorImg

plt.figure(figsize=(15,10))

plt.subplot(1,2,1)
plt.imshow(colorImg)
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(negColorImg)
plt.axis('off')

plt.show()
```
