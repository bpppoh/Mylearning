```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def imgShow(img) :
    cv2.imshow("image",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    for i in range(1) :
        cv2.waitKey(1)
```

```python
img = cv2.imread('../image/dark.jpg',0)
imgShow(img)
```

```python
gamma = 0.3
gammaTable = np.array([i/255.0 for i in range(256)])
print(gammaTable)
gammaTable = np.power(gammaTable,gamma)
print(gammaTable)
gammaTable = np.uint8(gammaTable*255)
print(gammaTable)
```
