```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../image/lena.jpg',0)
plt.imshow(img , cmap='gray')
plt.axis('off')
plt.show()
```

```python
out = np.zeros_like(img)
out[img>=64] = 255
plt.imshow(out , cmap='gray')
plt.axis('off')
plt.show()
```

```python
def bit_plane_slicing(img,bit) :
    bitImg = img.copy()
    bitImg = np.right_shift(bitImg,bit)
    bitImg = np.bitwise_and(bitImg,1)
    bitImg = (bitImg * 255).astype(np.uint8)
    print(f"operate at bit {bit} , {2**bit}")
    return bitImg
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(3,3,1)
plt.imshow(img , cmap='gray')
plt.axis('off')
for i in range(8):
    plt.subplot(3, 3, i+2)
    plt.imshow(bit_plane_slicing(img,i), cmap='gray')
    plt.title(f'Bit plane {i}')
    plt.xticks([])
    plt.yticks([])
plt.show()
```
